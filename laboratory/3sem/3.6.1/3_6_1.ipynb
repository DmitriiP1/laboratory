```python
import numpy as np
import scipy.stats as stats
import scipy.optimize as optim
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
tau = np.array([40, 57.8, 75.6, 93.3, 111.1, 128.9, 146.7, 164.4, 182.2, 200])
dv = np.array([25, 17, 13, 11, 9, 8, 7, 6, 5.5, 5])
sigma_tau = 0.4
sigma_dv = 0.4
```

```python
' & '.join([str(t) for t in tau])
```

```python
plt.figure(figsize=(14, 14))
X = np.linspace(40, 200, 5)
slope, intercept, r_value, p_value, std_err = stats.linregress(1/tau, dv)

plt.plot(1/X, slope / X + intercept, color='black', 
         label='y = {0:.4}*x + ({1:.4}), std={2:.4}'.format(slope, intercept, std_err))
plt.errorbar(1/tau, dv, sigma_dv, sigma_tau / tau / tau, color='gray', linestyle='None')
plt.grid()
plt.xlabel('1/tau, mcs^-1')
plt.ylabel('dv, kHz')
plt.title('Зависимость ширины спектра от длительности импульса')
plt.legend()
plt.savefig('data/3.6.1/img/1.png', format='png')
plt.show()
```
